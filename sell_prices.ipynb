```python
import pandas as pd
import numpy as np
```

```python
sell_prices = pd.read_csv('sell_prices.csv')
model = pd.read_csv('calendar.csv')
```

```python
model = model[model.columns[:2]]
```

```python
sell_prices_CA_1 = sell_prices.loc[sell_prices.store_id=='CA_1']
```

```python
sell_prices_CA_1['cat'] = np.nan
```

```python
sell_prices_CA_1
```

```python
sell_prices_CA_1['id'] = sell_prices_CA_1.item_id + '_' + sell_prices_CA_1.store_id +'_valildation'
```

```python
sell_prices_CA_1 = sell_prices_CA_1.drop('cat',axis=1)
```

```python
sell_prices_CA_1
```

```python
goods = sell_prices_CA_1.id.unique()
```

```python
for good in goods:
    model[good] = np.nan
```

```python
model
```

```python
sell_prices_CA_1
```

```python
for index,row in sell_prices_CA_1.iterrows():
    model.loc[model.wm_yr_wk==row.wm_yr_wk,row.id] = row.sell_price
```

```python
model = model.fillna(0)
```

```python
model.to_csv('prices_CA.csv')
```
