```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
```

```python
data = pd.read_csv('sales_train_validation.csv')
stores_CA = data.loc[data.state_id == 'CA'].store_id.unique()
store = stores_CA[0]
data_store = data.loc[data.store_id == store]
cats = data_store.cat_id.unique()
cat = cats[0]
data_store_cat = data_store.loc[data_store.cat_id == cat]
data_sales = data_store_cat[data_store_cat.columns[6:]].T
data_sales.columns = data_store_cat.id
```

```python
data_sales
```

```python
columns = data_sales.columns
shape = data_sales.shape[1]
```

```python
corr = pd.DataFrame(data=np.full((shape, shape), np.nan),
                        columns=columns, index=columns)
```

```python
corr
```

```python
for col1 in columns:
    for col2 in columns:
        corr.loc[col1, col2] = pearsonr(
            data_sales[col1], data_sales[col2])[0]
```

```python
corr
```

```python
corr.to_csv('corr.csv')
```
